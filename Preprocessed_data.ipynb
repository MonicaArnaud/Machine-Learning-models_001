**Read the data again**

```python
df = pd.read_csv('Land_Price_preprocessed.csv')
df.head(10)
```

```python
df.info()
```

```python
df.describe(include='all')
```

**Get dummies for Lease duration**

```python
Lease_duration_columns = pd.get_dummies(df['Lease duration'])
Lease_duration_columns.head(50)
```

```python
#Check the dummies
Lease_duration_columns['check']=Lease_duration_columns.sum(axis=1)
Lease_duration_columns
```

```python
#Rename the columns
column_names_a = ['Freehold','Leasehold','Unidentified','check']
Lease_duration_columns.columns = column_names_a
Lease_duration_columns.head(10)
```

```python
#Drop the "check" columns
Lease_duration_columns.drop(['check'],axis=1)
Lease_duration_columns.head()
```

**Get dummies for property type**

```python
property_type = pd.get_dummies(df['Property type'])
property_type.head(10)
```

```python
column_names = ['Detached','Flats/Maisonettes','Other','Semi-Detached','Terraced']
property_type.columns = column_names
property_type.head(15)
```

**Get dummies for Location column**

```python
#transform column "location" to two dummies, the fist is "London", the second is "Ohter cities"
def func_series_apply(london):
    if london == 'LONDON':
        return 'LONDON'
    else:
        return 'Other Cities'
    
df['Location']=df['Location'].apply(func_series_apply)
```

```python
df.head()
```

```python
Location_mod = pd.get_dummies(df['Location'])
Location_mod.head()
```

**Concatenate the dummies columns**

```python
data_processed = pd.concat([df,Location_mod,Lease_duration_columns,property_type],axis=1)
data_processed.head()
```

```python
#Remove the columns
data_processed = data_processed.drop(['Property type','Lease duration','Location','check'],axis=1)
data_processed.head()
```

**Normalize the date**

```python
data_processed['Date'] = pd.to_datetime(data_processed['Date'])
data_processed.info()
```

```python
data_processed['Date'] = pd.to_datetime(data_processed['Date']).dt.normalize()
data_processed.head()
```

**Final Checkpoint**

```python
data_processed_mod = data_processed.copy()
```

**Save the preprocessed data**

```python
data_processed_mod.to_csv('Data_processed_mod.csv',index=False)
```
